```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def compute_VSP(speed, acceleration, road_grade):
    
    gravity = 9.81
    rolling_resistance = 0.132
    aerodynamic_drag = 0.000302

    vehicle_specific_power = (speed * (1.1 * acceleration + gravity * np.sin(road_grade) + rolling_resistance) + aerodynamic_drag * speed ** 3)
    return vehicle_specific_power
```

### Acceleration fixed and road grade set to 0 -> increasing speed

```python
road_grade = 0.0
speed = np.linspace(0, 30, 100)
acceleration_x, acceleration_y, acceleration_z = (0.2, 0.2, 0.2)
acceleration = np.sqrt(acceleration_x ** 2 + acceleration_y ** 2 + acceleration_z ** 2)

for velocity in speed:
    vsp = compute_VSP(velocity, acceleration, road_grade)
    plt.plot(velocity, vsp, 'ro')

plt.xlabel('Speed [m/s]')
plt.ylabel('VSP [W]')
plt.title('Vehicle Specific Power in relation with increasing speed (other parameters are fixed)')
plt.grid()
plt.savefig("../plots/vsp/increasing_speed.png")
plt.show()
```

### increasing acceleration

```python
road_grade = 0.0
speed = 0.5

acceleration_x = np.linspace(0, 1, 100)
acceleration_y = np.linspace(0, 1, 100)
acceleration_z = np.linspace(0, 1, 100)

for x, y, z in zip(acceleration_x, acceleration_y, acceleration_z):
    acceleration = np.sqrt(x ** 2 + y ** 2 + z ** 2)
    vsp = compute_VSP(speed, acceleration, road_grade)
    plt.plot(acceleration, vsp, 'ro')

plt.xlabel('acceleration along each dimension [m/s^2]')
plt.ylabel('VSP [W]')
plt.title('Vehicle Specific Power in relation with increasing speed')
plt.grid()
plt.savefig("../plots/vsp/increasing_acceleration.png")
plt.show()
```

### increasing road grade

```python
speed = 0.5
acceleration_x, acceleration_y, acceleration_z = (0.2, 0.2, 0.2)
acceleration = np.sqrt(acceleration_x ** 2 + acceleration_y ** 2 + acceleration_z ** 2)

road_grade = np.linspace(0.0, 1.0, 1000)

for grade in road_grade:
    vsp = compute_VSP(speed, acceleration, grade)
    plt.plot(grade, vsp, 'ro')

plt.xlabel('Road Grade [rad]')
plt.ylabel('VSP [W]')
plt.title('Vehicle Specific Power in relation with increasing road grade (other parameters are fixed)')
plt.grid()
plt.savefig("../plots/vsp/increasing_grade.png")
```
